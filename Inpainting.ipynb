```python
import cv2
import matplotlib.pyplot as plt
```

```python
imageSet = ["airplane","doggo","kid","lena","messi","sky",]
# imageSet = ["airplane","doggo","kid","lena","messi","puzzle","sky",]
mountPt = "/content/drive/MyDrive/ivp_data/"
```

```python
def inpaint(imageName,key_):
    img = cv2.imread(mountPt+'images/'+imageName+'.jpg')
    mask = cv2.imread(mountPt+'masks/'+imageName+'.jpg',0)

    img_tel = cv2.inpaint(img,mask,3,cv2.INPAINT_TELEA)
    img_ns = cv2.inpaint(img,mask,3,cv2.INPAINT_NS)

    plt.subplot(6,4,key_*4+1)
    if key_==0 : plt.title("Damaged Image")
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

    plt.subplot(6,4,key_*4+2)
    if key_==0 : plt.title("Mask")
    plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))

    plt.subplot(6,4,key_*4+3)
    if key_==0 : plt.title("TELEA inpainting")
    plt.imshow(cv2.cvtColor(img_tel, cv2.COLOR_BGR2RGB))

    plt.subplot(6,4,key_*4+4)
    if key_==0 : plt.title("Navier Stokes inpainting")
    plt.imshow(cv2.cvtColor(img_ns, cv2.COLOR_BGR2RGB))
```

```python
plt.figure(figsize=(25, 25))

for x in range(len(imageSet)):
    inpaint(imageSet[x],x)
```
